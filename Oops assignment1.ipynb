````python
1. What are the five key concepts of Object-Oriented Programming (OOP)?

The five key concepts of Object-Oriented Programming (OOP) are:

1. **Encapsulation**:
   - Encapsulation is the concept of bundling data (attributes) and methods (functions) that operate on that data into a single unit, called a class.
   - It helps to restrict direct access to some of the object's components, which is known as **data hiding**. This ensures that the object's internal state can only be modified through its methods, improving control and security.

2. **Abstraction**:
   - Abstraction is the process of hiding the complex implementation details and showing only the essential features of an object.
   - It allows programmers to focus on a simplified view of an object or system, making the interaction with objects easier and reducing complexity.

3. **Inheritance**:
   - Inheritance is the mechanism by which one class (called a **subclass** or **derived class**) can inherit attributes and methods from another class (called a **superclass** or **base class**).
   - This allows for code reuse and the creation of hierarchies of classes, where more specialized behavior is built upon more general behavior.

4. **Polymorphism**:
   - Polymorphism allows objects of different classes to be treated as objects of a common superclass. It enables a single function or method to behave differently depending on the object it is acting upon.
   - This can be achieved through **method overriding** (inherited methods with modified behavior) and **method overloading** (same method name with different parameters).

5. **Composition**:
   - Composition refers to building complex objects by combining simpler objects. Rather than inheriting from a parent class, one object can contain references to other objects, thus forming a "has-a" relationship rather than an "is-a" relationship (which is typical of inheritance).
   - It promotes modularity and flexibility in the design of systems, where different components can be reused across different objects.

These core concepts provide a foundation for creating flexible, reusable, and maintainable code in object-oriented programming.

2. Write a Python class for a `Car` with attributes for `make`, `model`, and `year`. Include a method to display the car's information.

Here’s a Python class for a `Car` with attributes for `make`, `model`, and `year`, along with a method to display the car's information:

```python
class Car:
    # Constructor to initialize the car's attributes
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    # Method to display the car's information
    def display_info(self):
        print(f"Car Information:\nMake: {self.make}\nModel: {self.model}\nYear: {self.year}")

# Example of creating an instance of the Car class and displaying its information
my_car = Car("Toyota", "Corolla", 2020)
my_car.display_info()
```

### Explanation:
- **`__init__` method**: This is the constructor that initializes the car’s `make`, `model`, and `year` attributes when a new object is created.
- **`display_info` method**: This method prints the car's information (make, model, and year) in a readable format.

### Example Output:
```
Car Information:
Make: Toyota
Model: Corolla
Year: 2020
```

You can create instances of the `Car` class and use the `display_info` method to view the details of each car.

3. Explain the difference between instance methods and class methods. Provide an example of each.

In Python, **instance methods** and **class methods** are both types of methods associated with classes, but they differ in how they are defined and accessed.

### 1. **Instance Methods**:
   - **Definition**: Instance methods are functions that belong to instances (objects) of a class. They can access and modify the object's attributes and are called on an instance of the class.
   - **Self Parameter**: The first parameter of an instance method is always `self`, which refers to the current object (instance) of the class.
   - **Access**: Instance methods can access instance variables (attributes) and other methods in the class.

#### Example of an Instance Method:
```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    # Instance method
    def display_info(self):
        print(f"Car Info: {self.year} {self.make} {self.model}")

# Creating an instance of Car
my_car = Car("Toyota", "Corolla", 2020)
my_car.display_info()  # Calling the instance method
```

### Output:
```
Car Info: 2020 Toyota Corolla
```

In this example, `display_info` is an **instance method** because it accesses the instance's attributes (`self.make`, `self.model`, `self.year`).

### 2. **Class Methods**:
   - **Definition**: Class methods are functions that belong to the class itself rather than to an instance. They are called on the class, not an instance of the class.
   - **cls Parameter**: The first parameter of a class method is `cls`, which refers to the class itself, not an instance.
   - **Access**: Class methods can access class-level attributes (those defined at the class level) but not instance-level attributes directly. They are often used to modify class-level state or perform operations related to the class itself, rather than specific instances.

#### Example of a Class Method:
```python
class Car:
    # Class attribute
    total_cars = 0

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        # Increment the total number of cars whenever a new car is created
        Car.total_cars += 1

    # Class method
    @classmethod
    def get_total_cars(cls):
        print(f"Total cars created: {cls.total_cars}")

# Creating instances of Car
car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("Honda", "Civic", 2021)

# Calling the class method
Car.get_total_cars()  # Calling class method on the class itself
```

### Output:
```
Total cars created: 2
```

In this example, `get_total_cars` is a **class method** because it operates on the class (`cls.total_cars`) rather than an instance. The `total_cars` class attribute tracks how many `Car` objects have been created, and the class method is used to access and display that information.

### Key Differences:
- **Instance Methods**:
  - Operate on instances of the class.
  - The first parameter is `self`, which refers to the instance of the class.
  - Can access instance attributes.

- **Class Methods**:
  - Operate on the class itself, not on individual instances.
  - The first parameter is `cls`, which refers to the class.
  - Can access class attributes but not instance attributes directly.
  - Typically used for class-level operations (like keeping track of how many instances have been created).

4. How does Python implement method overloading? Give an example.

In Python, method overloading is not implemented in the same way as in other languages like Java or C++. Python does not support method overloading by default, meaning you cannot define multiple methods with the same name but different parameters. However, Python provides flexibility through default arguments, variable-length arguments, or using `*args` and `**kwargs` to simulate overloading.

### Method Overloading in Python Using Default Arguments

One common way to simulate method overloading in Python is by using default values for parameters, so that the function can accept different numbers of arguments.

#### Example:

```python
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

# Creating an instance of the Calculator class
calc = Calculator()

# Calling the add method with different numbers of arguments
print(calc.add(10))        # Output: 10 (only a is passed)
print(calc.add(10, 20))    # Output: 30 (a and b are passed)
print(calc.add(10, 20, 30)) # Output: 60 (a, b, and c are passed)
```

In this example, the `add` method is defined with default values for `b` and `c`. This allows us to call the method with either one, two, or three arguments, effectively simulating overloading.

### Method Overloading Using `*args` (Variable-Length Arguments)

You can also use `*args` to allow the method to accept any number of positional arguments, enabling overloading behavior.

#### Example:

```python
class Printer:
    def print_message(self, *args):
        for arg in args:
            print(arg)

# Creating an instance of the Printer class
printer = Printer()

# Calling the print_message method with different numbers of arguments
printer.print_message("Hello")           # Output: Hello
printer.print_message("Hello", "World")  # Output: Hello World
printer.print_message("Hello", "World", "Python")  # Output: Hello World Python
```

In this case, the `print_message` method can accept any number of arguments, allowing different behaviors depending on the number of parameters passed.

### Method Overloading Using `**kwargs` (Keyword Arguments)

`**kwargs` can be used to handle method overloading by accepting any number of keyword arguments.

#### Example:

```python
class Greet:
    def greet_person(self, **kwargs):
        if 'name' in kwargs:
            print(f"Hello, {kwargs['name']}!")
        else:
            print("Hello, Guest!")

# Creating an instance of the Greet class
greeter = Greet()

# Calling the greet_person method with or without a 'name' keyword
greeter.greet_person(name="John")  # Output: Hello, John!
greeter.greet_person()             # Output: Hello, Guest!
```

In this example, the `greet_person` method accepts any number of keyword arguments, and if a `name` is provided, it greets that person.

### Conclusion

Although Python does not support method overloading in the traditional sense (like C++ or Java), you can achieve similar behavior using default parameters, variable-length arguments (`*args`), or keyword arguments (`**kwargs`). This allows a method to handle a flexible number of inputs and provide different functionality based on the arguments passed.

5. What are the three types of access modifiers in Python? How are they denoted?

In Python, **access modifiers** are used to control the visibility or accessibility of variables, methods, and attributes within classes. Python follows a convention-based approach rather than enforcing strict access control like some other programming languages (such as Java or C++). The three types of access modifiers in Python are:

### 1. **Public Access Modifier**
   - **Denotation:** No underscores (`attribute` or `method`)
   - **Description:** Public members are accessible from outside the class. By default, all attributes and methods in a Python class are public, meaning they can be accessed directly using an instance of the class or by inheritance.

   #### Example:
   ```python
   class Person:
       def __init__(self, name):
           self.name = name  # Public attribute

       def greet(self):  # Public method
           print(f"Hello, {self.name}!")

   p = Person("Alice")
   print(p.name)  # Accessing the public attribute
   p.greet()      # Calling the public method
   ```

### 2. **Protected Access Modifier**
   - **Denotation:** A single leading underscore (`_attribute` or `_method`)
   - **Description:** Protected members are intended to be accessible only within the class and by subclasses (i.e., they are meant for internal use). However, this is only a convention and doesn't enforce strict access control. The use of a leading underscore signals to developers that the attribute or method is "protected" and should not be accessed directly from outside the class or subclass.

   #### Example:
   ```python
   class Person:
       def __init__(self, name):
           self._name = name  # Protected attribute

       def _greet(self):  # Protected method
           print(f"Hello from _greet, {self._name}!")

   class Employee(Person):
       def greet(self):
           print(f"Employee greeting, {self._name}!")

   p = Person("Alice")
   # Even though it's possible, it's not recommended to access protected attributes
   print(p._name)  # Accessing protected attribute directly
   p._greet()      # Calling protected method directly

   e = Employee("Bob")
   e.greet()  # Accessing protected attribute from subclass
   ```

### 3. **Private Access Modifier**
   - **Denotation:** A double leading underscore (`__attribute` or `__method`)
   - **Description:** Private members are intended to be used only within the class and are not accessible from outside the class, including from subclasses. Python does not prevent access to private members, but it "mangles" their names to make direct access harder. This is primarily a convention to indicate that the member is private and should not be accessed outside the class.

   - **Name Mangling:** Python changes the name of private attributes and methods by adding a prefix `_ClassName`, making them harder (but not impossible) to access from outside the class.

   #### Example:
   ```python
   class Person:
       def __init__(self, name):
           self.__name = name  # Private attribute

       def __greet(self):  # Private method
           print(f"Hello from __greet, {self.__name}!")

   p = Person("Alice")
   # Direct access to private members is not allowed
   # print(p.__name)    # This would raise an AttributeError
   # p.__greet()        # This would raise an AttributeError

   # However, you can access them using name mangling (not recommended)
   print(p._Person__name)  # Accessing the private attribute (name mangling)
   p._Person__greet()      # Calling the private method (name mangling)
   ```

### Summary:
- **Public (`attribute` or `method`)**: No leading underscores, accessible everywhere.
- **Protected (`_attribute` or `_method`)**: A single leading underscore, intended for use within the class and its subclasses.
- **Private (`__attribute` or `__method`)**: Double leading underscores, intended to be accessible only within the class itself (name mangling occurs).

Although Python uses these conventions, they are not enforced by the language, and it's possible to access "protected" or "private" members if necessary. The conventions serve as a guideline to promote good coding practices.

6. Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.

In Python, **inheritance** allows a class to inherit attributes and methods from another class, enabling code reuse and establishing a relationship between the parent and child classes. Python supports different types of inheritance, each allowing for different ways of extending and modifying behavior.

### 1. **Single Inheritance**
In **single inheritance**, a child class inherits from only one parent class. This is the simplest form of inheritance.

#### Example:
```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
dog = Dog()
dog.speak()  # Inherited method from Animal class
dog.bark()   # Method from Dog class
```

### 2. **Multiple Inheritance**
In **multiple inheritance**, a child class inherits from more than one parent class. This allows the child class to inherit attributes and methods from multiple classes, but it can introduce complexity if the classes have conflicting methods or attributes.

#### Example:
```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Mammal:
    def breathe(self):
        print("Mammal breathes")

class Dog(Animal, Mammal):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
dog = Dog()
dog.speak()   # Inherited from Animal class
dog.breathe() # Inherited from Mammal class
dog.bark()    # Defined in Dog class
```

In this example, `Dog` inherits from both `Animal` and `Mammal`, meaning it has access to methods from both parent classes.

### 3. **Multilevel Inheritance**
In **multilevel inheritance**, a class is derived from a class that is also derived from another class, forming a "chain" of inheritance.

#### Example:
```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Mammal(Animal):
    def breathe(self):
        print("Mammal breathes")

class Dog(Mammal):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
dog = Dog()
dog.speak()   # Inherited from Animal
dog.breathe() # Inherited from Mammal
dog.bark()    # Defined in Dog
```

In this case, `Dog` inherits from `Mammal`, which in turn inherits from `Animal`, forming a multilevel inheritance structure.

### 4. **Hierarchical Inheritance**
In **hierarchical inheritance**, multiple child classes inherit from a single parent class. This allows the parent class to be shared across several subclasses.

#### Example:
```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Creating instances of Dog and Cat
dog = Dog()
cat = Cat()

dog.speak()  # Inherited from Animal
cat.speak()  # Inherited from Animal
```

Here, both `Dog` and `Cat` inherit from the `Animal` class, forming a hierarchical inheritance structure.

### 5. **Hybrid Inheritance**
**Hybrid inheritance** is a combination of two or more types of inheritance. For example, it might combine multiple and multilevel inheritance, leading to more complex inheritance structures.

#### Example:
```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Mammal(Animal):
    def breathe(self):
        print("Mammal breathes")

class Bird(Animal):
    def fly(self):
        print("Bird flies")

class Bat(Mammal, Bird):
    def hang(self):
        print("Bat hangs upside down")

# Creating an instance of Bat
bat = Bat()
bat.speak()   # Inherited from Animal
bat.breathe() # Inherited from Mammal
bat.fly()     # Inherited from Bird
bat.hang()    # Defined in Bat
```

In this example, `Bat` inherits from both `Mammal` and `Bird`, creating a hybrid inheritance structure.

### Summary of Inheritance Types:
1. **Single Inheritance**: A child class inherits from one parent class.
2. **Multiple Inheritance**: A child class inherits from multiple parent classes.
3. **Multilevel Inheritance**: A class inherits from a class that is already a subclass of another class.
4. **Hierarchical Inheritance**: Multiple child classes inherit from a single parent class.
5. **Hybrid Inheritance**: A combination of two or more types of inheritance (e.g., multiple and multilevel inheritance).

### Simple Example of Multiple Inheritance:
As shown above, the **multiple inheritance** example combines functionality from two parent classes (`Animal` and `Mammal`) into a single child class (`Dog`):

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Mammal:
    def breathe(self):
        print("Mammal breathes")

class Dog(Animal, Mammal):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
dog = Dog()
dog.speak()   # Inherited from Animal class
dog.breathe() # Inherited from Mammal class
dog.bark()    # Defined in Dog class
```

This example demonstrates how the `Dog` class can inherit methods from both `Animal` and `Mammal` classes, making it an example of **multiple inheritance**.

7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

### Method Resolution Order (MRO) in Python

In Python, **Method Resolution Order (MRO)** refers to the order in which methods are inherited from a class hierarchy when a method is called. When a class inherits from multiple classes (as in **multiple inheritance**), Python follows a specific order to determine which method or attribute to use. The MRO is crucial to understanding how Python resolves conflicts when a method or attribute is present in more than one class in the inheritance chain.

The MRO ensures that the method lookup is consistent and predictable in cases of complex inheritance structures. Python follows the **C3 Linearization** algorithm to determine the MRO, which produces a consistent order of classes from which methods are inherited.

### MRO and the C3 Linearization

Python's MRO follows the **C3 Linearization** (C3 superclass linearization) algorithm. This algorithm determines the order in which classes are considered during method lookup in a multiple inheritance situation. The goal of the C3 algorithm is to ensure that:
1. The base class methods are resolved first.
2. It respects the inheritance hierarchy and resolves conflicts in a consistent order.
3. It does not break any parent-child relationships.

### How MRO Works

For example, consider the following class structure:

```python
class A:
    def method(self):
        print("A's method")

class B(A):
    def method(self):
        print("B's method")

class C(A):
    def method(self):
        print("C's method")

class D(B, C):
    pass
```

In this case, `D` inherits from both `B` and `C`. When calling the `method` of a `D` instance, Python needs to decide which `method` to call, and it uses the MRO to determine that order.

```python
d = D()
d.method()  # Which method will be called?
```

To determine the order, Python follows the MRO, which in this case will be: `D -> B -> C -> A`.

### Retrieving the MRO Programmatically

To retrieve the MRO programmatically, you can use the `mro()` method or the `__mro__` attribute, both of which return the method resolution order of a class.

#### Using `mro()` method
The `mro()` method returns a list of classes in the MRO for a given class.

```python
print(D.mro())  # Using the mro() method to get the MRO
```

#### Using `__mro__` attribute
You can also access the `__mro__` attribute directly to get the MRO.

```python
print(D.__mro__)  # Accessing the __mro__ attribute
```

### Example:

```python
class A:
    def method(self):
        print("A's method")

class B(A):
    def method(self):
        print("B's method")

class C(A):
    def method(self):
        print("C's method")

class D(B, C):
    pass

# Printing the MRO using mro() and __mro__
print(D.mro())  # Using the mro() method
print(D.__mro__)  # Using the __mro__ attribute
```

#### Output:

```python
[D, B, C, A, <class 'object'>]
[D, B, C, A, <class 'object'>]
```

Both `mro()` and `__mro__` return the same result. The MRO for the class `D` is: `D -> B -> C -> A -> object`. This means when you call a method on an instance of `D`, Python will first look for the method in `D`, then `B`, followed by `C`, `A`, and finally the `object` class (which is the base class for all Python classes).

### Summary:

- **MRO (Method Resolution Order)** in Python is the order in which methods are inherited from the class hierarchy during method lookup.
- It is especially important in **multiple inheritance** scenarios to determine which class's method will be called.
- Python uses the **C3 Linearization** algorithm to compute the MRO.
- You can retrieve the MRO programmatically using:
  - `ClassName.mro()` method
  - `ClassName.__mro__` attribute

This ensures a consistent and predictable method lookup order, even in complex inheritance chains.

8. Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses `Circle` and `Rectangle` that implement the `area()` method.

In Python, an **abstract base class (ABC)** is defined using the `abc` module. An **abstract method** is a method that is declared in an abstract class but contains no implementation in that class. The subclasses are required to implement this abstract method.

### Steps:
1. **Define the abstract base class `Shape`** with an abstract method `area()`.
2. **Create two subclasses `Circle` and `Rectangle`**, each implementing the `area()` method.

Here's how you can do this:

```python
from abc import ABC, abstractmethod
import math

# Abstract Base Class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass  # Abstract method, no implementation

# Subclass Circle that implements the area() method
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Subclass Rectangle that implements the area() method
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage:
circle = Circle(5)
print(f"Area of the circle: {circle.area()}")  # Output: Area of the circle: 78.53981633974483

rectangle = Rectangle(4, 6)
print(f"Area of the rectangle: {rectangle.area()}")  # Output: Area of the rectangle: 24
```

### Explanation:
1. **Abstract Base Class (`Shape`)**:
   - `Shape` is an abstract base class derived from `ABC`.
   - The method `area()` is marked with `@abstractmethod`, meaning any class that inherits from `Shape` must implement this method.

2. **Subclasses (`Circle` and `Rectangle`)**:
   - Both `Circle` and `Rectangle` are concrete subclasses of `Shape`.
   - They implement the `area()` method, which calculates the area of the respective shape:
     - `Circle`: The area is calculated using the formula \( \pi \times \text{radius}^2 \).
     - `Rectangle`: The area is calculated using the formula \( \text{width} \times \text{height} \).

3. **Instantiating and Using the Subclasses**:
   - You can create instances of `Circle` and `Rectangle`, and call their `area()` method to get the area.

### Output:
```
Area of the circle: 78.53981633974483
Area of the rectangle: 24
```

This demonstrates the use of an abstract base class and its implementation in subclasses. The abstract method `area()` ensures that all subclasses must provide their own implementation of how to calculate the area.

9. Demonstrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.
Polymorphism in Python allows objects of different classes to be treated as objects of a common base class. Specifically, polymorphism enables a function to work with objects of different types (i.e., different classes) and use their methods in a common way. In this case, we can create a function that can accept different shape objects (like `Circle` and `Rectangle`) and call their respective `area()` method to calculate and print the area, demonstrating polymorphism.

### Steps:
1. Create a function that can accept any object that is a subclass of the `Shape` class.
2. The function will call the `area()` method, which is implemented differently in each subclass.

Here's the implementation:

```python
from abc import ABC, abstractmethod
import math

# Abstract Base Class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass  # Abstract method, no implementation

# Subclass Circle that implements the area() method
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Subclass Rectangle that implements the area() method
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Function to calculate and print the area of any shape
def print_area(shape: Shape):
    print(f"The area of the shape is: {shape.area()}")

# Example usage:
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Demonstrating polymorphism by passing different shape objects to the same function
print_area(circle)       # Polymorphism in action (Circle)
print_area(rectangle)    # Polymorphism in action (Rectangle)
```

### Explanation:
1. **Polymorphism in Action**:
   - The function `print_area()` is designed to accept any object that is a subclass of `Shape` (i.e., it can accept both `Circle` and `Rectangle`).
   - It calls the `area()` method on the passed object, which, depending on the object's class, will execute the respective implementation of `area()` (either in `Circle` or `Rectangle`).

2. **The `area()` Method**:
   - Each shape (`Circle` and `Rectangle`) has its own version of the `area()` method.
     - For `Circle`, it calculates the area using the formula \( \pi \times \text{radius}^2 \).
     - For `Rectangle`, it calculates the area using the formula \( \text{width} \times \text{height} \).

3. **Demonstrating Polymorphism**:
   - When `print_area(circle)` is called, it calls the `area()` method of `Circle`, and when `print_area(rectangle)` is called, it calls the `area()` method of `Rectangle`.
   - Despite being different shapes, both objects are passed to the same function, and the function works correctly due to polymorphism.

### Output:
```
The area of the shape is: 78.53981633974483  # Area of the circle
The area of the shape is: 24                    # Area of the rectangle
```

### Summary:
In this example, polymorphism allows the `print_area()` function to work with different shape objects (`Circle` and `Rectangle`). The function doesn't need to know the specifics of each class, it simply calls the `area()` method, which behaves differently depending on the object type. This demonstrates how Python's polymorphism works in practice: the same method name (`area()`) can be used across different classes with different implementations.

10. Implement encapsulation in a `BankAccount` class with private attributes for `balance` and `account_number`. Include methods for deposit, withdrawal, and balance inquiry.

   To implement encapsulation in a `BankAccount` class with private attributes for `balance` and `account_number`, you can use Python's name mangling feature to make these attributes private. The private attributes will be accessed and modified through getter and setter methods, and methods for deposit, withdrawal, and balance inquiry will provide controlled access to them.

Here’s how to implement it:

```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        # Private attributes
        self.__account_number = account_number
        self.__balance = balance

    # Getter method for account number
    def get_account_number(self):
        return self.__account_number

    # Getter method for balance
    def get_balance(self):
        return self.__balance

    # Deposit method
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    # Withdrawal method
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        elif amount > self.__balance:
            print("Insufficient balance for the withdrawal.")
        else:
            print("Withdrawal amount must be positive.")

    # Balance inquiry method
    def inquire_balance(self):
        return f"Current balance: ${self.__balance}"

# Example usage:
account = BankAccount("123456789", 500)
print(account.inquire_balance())  # Should show balance of 500
account.deposit(150)  # Deposits 150
account.withdraw(200)  # Withdraws 200
print(account.get_account_number())  # Access the account number using the getter method
```

### Key Points:
1. **Private Attributes**: `self.__account_number` and `self.__balance` are private attributes (denoted by the double underscores `__`). They cannot be accessed directly from outside the class.
2. **Getter Methods**: Methods like `get_account_number` and `get_balance` are used to safely access these private attributes from outside the class.
3. **Deposit Method**: The `deposit` method ensures that the deposit amount is positive before adding it to the balance.
4. **Withdrawal Method**: The `withdraw` method checks if the withdrawal amount is valid (positive and does not exceed the available balance).
5. **Balance Inquiry**: The `inquire_balance` method allows you to check the current balance without modifying it.

This encapsulation pattern helps ensure that the balance and account number cannot be modified directly from outside the class, providing better control over the data.

11. Write a class that overrides the `__str__` and `__add__` magic methods. What will these methods allow you to do?

    In Python, you can override special methods (magic methods) to customize the behavior of your objects. The `__str__` method is used to define how an object is represented as a string when you use `str()` or print the object. The `__add__` method is used to define how the `+` operator behaves when applied to objects of your class.

Here’s an example of a class that overrides both the `__str__` and `__add__` methods:

### Example Implementation:

```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    # Overriding the __str__ method to customize string representation
    def __str__(self):
        return f"BankAccount(Account Number: {self.account_number}, Balance: ${self.balance})"

    # Overriding the __add__ method to define how the + operator works
    def __add__(self, other):
        if isinstance(other, BankAccount):
            # Adding the balances of two BankAccount objects
            new_balance = self.balance + other.balance
            # Creating a new BankAccount object with a new balance
            return BankAccount("MergedAccount", new_balance)
        return NotImplemented  # If the other object is not a BankAccount

# Example usage:
account1 = BankAccount("123456", 1000)
account2 = BankAccount("789012", 1500)

# Using the __str__ method for string representation
print(account1)  # Output: BankAccount(Account Number: 123456, Balance: $1000)

# Using the __add__ method (the + operator)
merged_account = account1 + account2  # This uses the __add__ method
print(merged_account)  # Output: BankAccount(Account Number: MergedAccount, Balance: $2500)
```

### Explanation of the Magic Methods:

1. **`__str__` method**:
   - The `__str__` method is used to provide a human-readable string representation of the object. This method is called when you try to convert the object to a string, for example, when you use `print()` or `str()`.
   - In this example, it returns a formatted string containing the account number and balance of the `BankAccount` object.
   - When you call `print(account1)`, the `__str__` method is automatically invoked, and you see the output in the format specified inside the method.

2. **`__add__` method**:
   - The `__add__` method allows you to define what happens when the `+` operator is used with objects of your class. It is automatically called when you do something like `object1 + object2`.
   - In this case, when two `BankAccount` objects are added, their balances are summed up, and a new `BankAccount` object is created with the new balance. The account number for the new object is set to `"MergedAccount"`.
   - When you do `account1 + account2`, Python internally calls `account1.__add__(account2)`.

### What Do These Methods Allow You to Do?

- **`__str__`**:
   - It allows you to control how your object is represented as a string. This is especially useful when you want a custom string format to display your object in a more user-friendly or readable way. Instead of printing memory addresses or default representations, the `__str__` method lets you define a descriptive string that summarizes the object’s state.
   - Example usage: `print(account1)` would print `BankAccount(Account Number: 123456, Balance: $1000)`.

- **`__add__`**:
   - It allows you to define how the `+` operator works for your objects. In this case, adding two `BankAccount` objects results in a new `BankAccount` object with a combined balance.
   - You can perform operations between two instances of the class using `+`, even though it would normally not make sense to add two objects together. You define the logic for what "adding" means within your class.
   - Example usage: `account1 + account2` would create a new `BankAccount` object with a combined balance of $2500.

In summary, overriding `__str__` provides a meaningful string representation, and overriding `__add__` allows you to control the behavior of the `+` operator between objects of your class.

12. Create a decorator that measures and prints the execution time of a function.

    To create a decorator that measures and prints the execution time of a function, we can make use of Python's `time` module. The decorator will wrap the target function, calculate the time before and after its execution, and then print the elapsed time.

Here’s how you can implement such a decorator:

### Implementation:

```python
import time

# Decorator to measure execution time
def measure_execution_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()  # Record the start time
        result = func(*args, **kwargs)  # Call the original function
        end_time = time.time()  # Record the end time

        execution_time = end_time - start_time  # Calculate the execution time
        print(f"Execution time of '{func.__name__}': {execution_time:.6f} seconds")

        return result  # Return the result of the original function

    return wrapper

# Example usage:

@measure_execution_time
def some_time_consuming_function():
    time.sleep(2)  # Simulate a time-consuming task (e.g., sleep for 2 seconds)

# Call the function
some_time_consuming_function()
```

### Explanation:

1. **`measure_execution_time` Decorator**:
   - This decorator takes a function (`func`) as an argument.
   - Inside the decorator, the `wrapper` function is defined, which wraps the original function.
   - Before calling the function, it records the start time using `time.time()`, which returns the current time in seconds since the epoch (as a float).
   - After the function executes, it records the end time and calculates the elapsed time by subtracting the start time from the end time.
   - The execution time is then printed in a readable format (with six decimal places).
   - Finally, the result of the original function is returned by the wrapper function.

2. **`some_time_consuming_function` Example**:
   - The function `some_time_consuming_function` simulates a time-consuming task by using `time.sleep(2)`, which pauses execution for 2 seconds.
   - When this function is called, the decorator measures and prints the time it took to execute.

### Output:

```
Execution time of 'some_time_consuming_function': 2.000123 seconds
```

### Key Points:
- **Decorator**: The `measure_execution_time` function is a decorator, which means it takes a function, wraps it with additional behavior (in this case, timing), and returns the wrapped function.
- **Time Measurement**: `time.time()` is used to capture the time before and after the function call to calculate the elapsed time.
- **Formatted Output**: The time is printed with six decimal places for precision, but you can adjust this based on your needs.

This approach allows you to easily measure the execution time of any function by simply decorating it with `@measure_execution_time`.


13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

   The **Diamond Problem** is a well-known issue in object-oriented programming, particularly in languages that support **multiple inheritance**. It occurs when a class inherits from two or more classes that have a common ancestor, creating a diamond-shaped inheritance structure.

### Diamond Problem Explained:

Consider the following inheritance structure:

```
        A
       / \
      B   C
       \ /
        D
```

- Class `A` is the base class.
- Class `B` and class `C` both inherit from `A`.
- Class `D` inherits from both `B` and `C`.

The problem arises when we try to instantiate `D`. The key issue is how Python (or any language with multiple inheritance) should handle method resolution when class `D` inherits from both `B` and `C`, and both `B` and `C` inherit from `A`. If `B` and `C` both override methods from `A`, and `D` does not override them, the question is: **Which version of the method from `A` should be called?** The method from `B` or the one from `C`?

### Problem Details:
- **Ambiguity**: There’s ambiguity in determining which class's method to use when a method from `A` is invoked on an instance of `D`.
- **Redundancy**: If `A`'s method is called indirectly through both `B` and `C`, you might get redundant or conflicting behavior.

### Python's Solution: Method Resolution Order (MRO)

Python resolves the Diamond Problem using a method called **Method Resolution Order (MRO)**. This approach ensures that the method resolution is predictable, consistent, and avoids ambiguity.

Python uses a technique called **C3 Linearization** (also known as C3 superclass linearization) to create a clear, well-defined order in which methods are inherited and called. The MRO defines the order in which classes are looked up when searching for a method.

### MRO in Python:
- The MRO is a linear ordering of classes that Python uses when searching for a method. The method is searched first in the class of the instance, then in its base classes in the order defined by the MRO.
- You can access the MRO of a class using the `mro()` method or `.__mro__` attribute.

#### Example:

```python
class A:
    def method(self):
        print("Method from A")

class B(A):
    def method(self):
        print("Method from B")

class C(A):
    def method(self):
        print("Method from C")

class D(B, C):
    pass

# Create an instance of D
d = D()

# Call the method
d.method()
```

#### Output:

```
Method from B
```

### Explanation:
1. **Class `D`** inherits from both `B` and `C`, and neither `B` nor `C` overrides the `method` in such a way that would create a conflict for `D`.
2. When `d.method()` is called, Python looks up the MRO for class `D`.
3. The MRO for class `D` is:
   ```
   D -> B -> C -> A
   ```
4. Since `B` is first in the MRO, the method from `B` is called, and the output is `"Method from B"`.

You can check the MRO for `D` by printing it:

```python
print(D.mro())
```

This will output:
```
[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

### How Python Resolves the Diamond Problem:
Python uses **C3 Linearization** to establish the MRO. This algorithm ensures that:
- The base class `A` appears only once in the MRO, even though it is indirectly inherited by both `B` and `C`.
- The MRO respects the inheritance order as specified in the class definition (i.e., `B` before `C` in the case of `D`).
- If two classes (`B` and `C`) both define the same method, the method from the class first in the MRO (`B` in this case) is used.

### In Summary:
- The **Diamond Problem** arises in multiple inheritance when a class inherits from two classes that share a common ancestor, leading to ambiguity in method resolution.
- **Python resolves this problem using the Method Resolution Order (MRO)**, which defines a clear, linear path for method lookups.
- **C3 Linearization** is used to create the MRO and ensures that classes are searched in a predictable order, avoiding ambiguity and redundancy in method resolution.

By using this approach, Python effectively handles the complexities of multiple inheritance while maintaining a clear and consistent inheritance structure.

14. Write a class method that keeps track of the number of instances created from a class.

    To keep track of the number of instances created from a class, we can define a **class variable** that will store the count of instances. Then, we can use a **class method** to access and update this count every time a new instance is created.

Here’s how you can implement it:

### Example Implementation:

```python
class MyClass:
    # Class variable to keep track of the number of instances
    instance_count = 0

    def __init__(self):
        # Increment the instance count each time a new object is created
        MyClass.instance_count += 1

    # Class method to access the instance count
    @classmethod
    def get_instance_count(cls):
        return cls.instance_count

# Example usage:
obj1 = MyClass()
obj2 = MyClass()
obj3 = MyClass()

# Get the number of instances created
print(MyClass.get_instance_count())  # Output: 3
```

### Explanation:
1. **Class Variable** (`instance_count`):
   - The class variable `instance_count` is shared among all instances of the class `MyClass`. It is used to keep track of the number of instances created from the class.

2. **Constructor** (`__init__`):
   - Every time a new instance of `MyClass` is created, the constructor `__init__` is called. Inside the constructor, we increment the `instance_count` by 1.

3. **Class Method** (`get_instance_count`):
   - The class method `get_instance_count` is defined with the `@classmethod` decorator. This method takes the class (`cls`) as the first argument and returns the current value of `instance_count`.
   - Class methods can be called on the class itself, not just on an instance of the class. This is why `MyClass.get_instance_count()` works, even though it's not tied to an instance.

4. **Example Usage**:
   - We create three instances of `MyClass` (obj1, obj2, obj3).
   - After creating these instances, the `instance_count` will be 3.
   - Calling `MyClass.get_instance_count()` will return 3, which is the total number of instances created.

### Output:
```
3
```

### Key Points:
- **Class Variable**: `instance_count` is shared by all instances of the class, and it is used to track how many instances have been created.
- **Class Method**: `get_instance_count` is a class method that can be called on the class itself to return the value of the `instance_count`.
- **Constructor**: Each time an instance is created, the constructor (`__init__`) updates the `instance_count`.

This approach ensures that you have a central, easy way to track how many objects are created from a class.

15. Implement a static method in a class that checks if a given year is a leap year.

    A **static method** in Python is a method that doesn't require access to the instance (`self`) or the class (`cls`). It's bound to the class rather than the instance, and it's often used for utility functions that perform some operation related to the class but don't need access to its instance attributes.

In this case, we can implement a static method that checks whether a given year is a **leap year** or not. According to the leap year rules:
- A year is a leap year if it is divisible by 4.
- However, if it is divisible by 100, it is not a leap year, unless it is also divisible by 400.

### Leap Year Check Implementation:

```python
class DateUtils:
    @staticmethod
    def is_leap_year(year):
        # Check if the year is divisible by 4, but not by 100, unless it's also divisible by 400
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        else:
            return False

# Example usage:
print(DateUtils.is_leap_year(2020))  # Output: True (2020 is a leap year)
print(DateUtils.is_leap_year(2023))  # Output: False (2023 is not a leap year)
print(DateUtils.is_leap_year(1900))  # Output: False (1900 is not a leap year)
print(DateUtils.is_leap_year(2000))  # Output: True (2000 is a leap year)
```

### Explanation:
1. **Static Method**:
   - The method `is_leap_year` is defined as a static method using the `@staticmethod` decorator. It does not take `self` or `cls` as the first parameter, because it does not need to access instance or class variables.

2. **Leap Year Logic**:
   - The method checks if the given year follows the leap year rules:
     - If the year is divisible by 4 and not by 100, or if it is divisible by 400, it returns `True` (indicating it's a leap year).
     - Otherwise, it returns `False` (indicating it's not a leap year).

3. **Example Usage**:
   - We test the method with several years to check if it's working correctly:
     - 2020 is divisible by 4 but not by 100, so it's a leap year.
     - 2023 is not divisible by 4, so it's not a leap year.
     - 1900 is divisible by 100 but not 400, so it's not a leap year.
     - 2000 is divisible by 400, so it's a leap year.

### Output:
```
True
False
False
True
```

### Key Points:
- **Static Method**: `is_leap_year` is defined as a static method because it does not need to interact with instance-specific data.
- **Leap Year Check**: The method checks whether the year meets the criteria for a leap year using modular arithmetic.
- **Utility Function**: Static methods are often used for utility functions like this one, where the function operates independently of any instance-specific data.

This implementation provides a clean and efficient way to check if a given year is a leap year using a static method.
````
